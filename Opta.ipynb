```python

import numpy as np
import matplotlib.pyplot as plt
from mplsoccer import Pitch
from sklearn.cluster import KMeans
import seaborn as sns

# Generate defender-like passes
np.random.seed(42)
num_passes = 300
start_coords = np.random.normal(loc=[30, 40], scale=[10, 20], size=(num_passes, 2))
end_coords = start_coords + np.random.normal(loc=[5, 0], scale=[10, 15], size=(num_passes, 2))
statuses = np.random.randint(0, 2, size=num_passes)

# Clip coordinates to stay within pitch boundaries
start_coords = np.clip(start_coords, [0, 0], [100, 80])
end_coords = np.clip(end_coords, [0, 0], [100, 80])

# Clustering pass start points
kmeans = KMeans(n_clusters=6, n_init=10)
kmeans.fit(start_coords)
clusters = kmeans.predict(start_coords)

# Create a pitch
pitch = Pitch(pitch_type='opta', pitch_color='#e0e0e0', line_color='black')
fig, ax = pitch.draw(figsize=(12, 8))

# Plotting heatmap for pass start locations
sns.kdeplot(x=start_coords[:, 0], y=start_coords[:, 1], fill=True, cmap='Blues', ax=ax, levels=100, thresh=0.01, alpha=0.6)

# Quiver plot for average pass directions from each cluster
for cluster_num in np.unique(clusters):
    cluster_indices = np.where(clusters == cluster_num)
    start_mean = np.mean(start_coords[cluster_indices], axis=0)
    end_mean = np.mean(end_coords[cluster_indices] - start_coords[cluster_indices], axis=0)
    ax.quiver(start_mean[0], start_mean[1], end_mean[0], end_mean[1], angles='xy', scale_units='xy', scale=2.5, color='red', width=1.5, headlength=4, headwidth=4)

# Statistics
successful_passes = np.sum(statuses)
unsuccessful_passes = num_passes - successful_passes
accuracy = successful_passes / num_passes * 100

# Display statistics
stats = f"Successful: {successful_passes}\nUnsuccessful: {unsuccessful_passes}\nTotal Passes: {num_passes}\nAccuracy: {accuracy:.2f}%"
props = dict(boxstyle='round', facecolor='white', alpha=0.8)
ax.text(5, 5, stats, verticalalignment='top', fontsize=12, bbox=props, fontweight='bold')

plt.title("Defender Pass Map with Heatmap & Directions", fontsize=20, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Data (Dummy)
minutes = np.arange(0, 106, 1)
cp_threat = [0, 0.5, 1, 1.5, 1, 0.5, 0, -0.5, -1, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 1, 0.5, 0, -1, -2, -2.5, -2, -1, 0, 1, 2, 2.5, 2, 1.5, 1, 0, -0.5, -1, -2, -2.5, -3, -2.5, -2, -1, 0, 1, 1.5, 1, 0.5, 0, 0, 0.5, 1, 1.5, 1, 0.5, 0]
cp_threat += cp_threat[::-1] # Mirror data for second half
cp_threat = np.array(cp_threat)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

# Plotting the data
ax.fill_between(minutes, cp_threat, 0, where=(cp_threat > 0), color='#E63946', interpolate=True, alpha=0.7, label='Crystal Palace more threatening')
ax.fill_between(minutes, cp_threat, 0, where=(cp_threat < 0), color='#457B9D', interpolate=True, alpha=0.7, label='Tottenham more threatening')

# Add soccer ball icons for goals
goal_times = [53, 66, 94]
for goal in goal_times:
    ax.text(goal, cp_threat[goal] + 0.5, '', ha='center', va='center', fontsize=16, color='black')
    ax.axvline(goal, color='grey', linestyle='--', alpha=0.5)

# Add sync icons for substitutes
sub_times = [30, 60, 75]
for sub in sub_times:
    ax.text(sub, cp_threat[sub] - 0.8, '', ha='center', va='center', fontsize=16, color='black')
    ax.axvline(sub, color='grey', linestyle='--', alpha=0.5)

# Add half-time line
ax.axvline(45, color='black', linestyle='-', alpha=0.7)
ax.text(45, 2.8, "HT", ha='center', va='center', fontsize=12, color='black', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.2'))

# Add title and labels
ax.set_title("Crystal Palace 1-2 Tottenham Hotspur\nPremier League 2023-24 | 27 October 2023", fontsize=16, fontweight='bold')
ax.set_ylabel("Attacking Threat", fontsize=14)
ax.set_xlabel("Minute", fontsize=14)
ax.legend(fontsize=12, loc='upper right')
ax.tick_params(labelsize=12)
ax.set_xlim(0, 105)
ax.set_ylim(-3.5, 3.5)

# Grid and background color
ax.set_facecolor('#FAFAFA')
ax.grid(axis='y', linestyle='-', linewidth=0.5, color='grey', alpha=0.3)

# Display the plot
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.legend_handler import HandlerTuple

# Data (Dummy)
minutes = np.arange(0, 106, 1)
cp_threat = [0, 0.5, 1, 1.5, 1, 0.5, 0, -0.5, -1, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 1, 0.5, 0, -1, -2, -2.5, -2, -1, 0, 1, 2, 2.5, 2, 1.5, 1, 0, -0.5, -1, -2, -2.5, -3, -2.5, -2, -1, 0, 1, 1.5, 1, 0.5, 0, 0, 0.5, 1, 1.5, 1, 0.5, 0]
cp_threat += cp_threat[::-1]
cp_threat = np.array(cp_threat)

fig, ax = plt.subplots(figsize=(14, 8))

# Plotting the data
ax.fill_between(minutes, cp_threat, 0, where=(cp_threat > 0), color='#E63946', interpolate=True, alpha=0.7)
ax.fill_between(minutes, cp_threat, 0, where=(cp_threat < 0), color='#457B9D', interpolate=True, alpha=0.7)

# Add soccer ball icons for goals
goal_times = [53, 66, 94]
for goal in goal_times:
    ax.text(goal, cp_threat[goal] + 0.5, '', ha='center', va='center', fontsize=16, color='black')
    ax.axvline(goal, color='grey', linestyle='--', alpha=0.5)
    ax.text(goal, 3, f"{goal}'", ha='center', va='center', fontsize=10, color='grey')

# Add sync icons for substitutes
sub_times = [30, 60, 75]
for sub in sub_times:
    ax.text(sub, cp_threat[sub] - 0.8, '', ha='center', va='center', fontsize=16, color='black')
    ax.axvline(sub, color='grey', linestyle='--', alpha=0.5)
    ax.text(sub, -3, f"{sub}'", ha='center', va='center', fontsize=10, color='grey')

# Half-time line
ax.axvline(45, color='black', linestyle='-', alpha=0.7)
ax.text(45, 2.8, "HT", ha='center', va='center', fontsize=12, color='black', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.2'))

# Title and labels
ax.set_title("Crystal Palace 1-2 Tottenham Hotspur\nPremier League 2023-24 | 27 October 2023", fontsize=16, fontweight='bold')
ax.set_ylabel("Attacking Threat", fontsize=14)
ax.set_xlabel("Minute", fontsize=14)
l1, = ax.fill(np.NaN, np.NaN, color='#E63946', alpha=0.7)
l2, = ax.fill(np.NaN, np.NaN, color='#457B9D', alpha=0.7)
ax.legend([(l1, l2)], ['Game Flow'], loc='upper right', handler_map={tuple: HandlerTuple(ndivide=None)}, fontsize=12)
ax.tick_params(labelsize=12)
ax.set_xlim(0, 105)
ax.set_ylim(-3.5, 3.5)

# Grid & background
ax.set_facecolor('#FAFAFA')
ax.grid(axis='y', linestyle='-', linewidth=0.5, color='grey', alpha=0.3)
ax.grid(axis='x', linestyle='-', linewidth=0.1, color='grey', alpha=0.2)

plt.tight_layout()
plt.show()
```
